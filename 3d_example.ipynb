```
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```
import numpy as np
```

```
n=100
```

```
x=np.linspace(-2,2,num=n)
y=np.linspace(-2,2,num=n)
x,y=np.meshgrid(x,y)
```

```
def fu(x,y,n):
  a=((x**n + y**n))**n
  return 1/(a-1)
def f(x,y):
  a=((x**2 + y**2))**2
  return 1/(a+1)
def fs(x,y):
  a=(np.cos(x)+np.sin(y))**len(x)
  return a
def cos(x):
  return np.cos(x)
```

```
fig=plt.figure(figsize=[20,15])
ax = fig.add_subplot(2, 2, 1, projection='3d')
ax.plot_surface(x,y,fu(x,y,n),cmap=cm.Blues)
ax.set_xlabel('x-axis',{'fontsize':25})
ax.set_ylabel('y-axis',{'fontsize':25})
ax.set_zlabel('z-axis',{'fontsize':25})
ax = fig.add_subplot(222, projection='3d')
ax.plot_surface(x,y,f(x,y),rstride=5, cstride=5,cmap='autumn')
ax.set_xlabel('x-axis',{'fontsize':25})
ax.set_ylabel('y-axis',{'fontsize':25})
ax.set_zlabel('z-axis',{'fontsize':25})
ax = fig.add_subplot(223, projection='3d')
ax.plot_surface(x,y,fs(x,y),cmap=cm.viridis)
ax.set_xlabel('x-axis',{'fontsize':25})
ax.set_ylabel('y-axis',{'fontsize':25})
ax.set_zlabel('z-axis',{'fontsize':25})
ax = fig.add_subplot(224, projection='3d')
ax.plot_surface(x,y,cos(x),rstride=1, cstride=1,cmap=cm.gist_earth)
ax.set_xlabel('x-axis',{'fontsize':25})
ax.set_ylabel('y-axis',{'fontsize':25})
ax.set_zlabel('z-axis',{'fontsize':25})
plt.show()
```
