```
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
import cv2
```

```
image = img.imread('test.jpg')
plt.imshow(image)
```

```
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```
blur_gray_image = cv2.GaussianBlur(gray_image,(5,5),0)
```

```
low = 125
high = 250
edge_image = cv2.Canny(blur_gray_image,low,high)
```

```
plt.imshow(edge_image,cmap='Greys_r')
```

```
rho = 1
theta = np.pi/180
threshold = 1
min_line_length = 10
max_line_gap = 1
copy = np.copy(image)
```

```
lines = cv2.HoughLinesP(edge_image, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)
#print(lines)
```

```
for line in lines:
  for w,x,y,z in line:
    cv2.line(edge_image,(w,x),(y,z),10)
plt.imshow(edge_image,cmap='gray')
```
