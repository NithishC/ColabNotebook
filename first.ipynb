```
data=pd.read_csv('credit_data.csv')
```

```
data.head()
```

```
from sklearn.model_selection import train_test_split
```

```
x=data.drop('default',axis=1)
y=data['default']
```

```
 X_train, X_test, y_train, y_test =train_test_split(x,y,test_size=0.3,random_state=101)
```

```
clientid=tf.feature_column.numeric_column("clientid")
income=tf.feature_column.numeric_column("income")
age=tf.feature_column.numeric_column("age")
loan=tf.feature_column.numeric_column("loan")
LTI=tf.feature_column.numeric_column("LTI")
```

```
feat_cols=[clientid,income,age,loan,LTI]
```

```
input_func=tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=100,num_epochs=None,shuffle=True)
```

```
model=tf.estimator.LinearClassifier(feature_columns=feat_cols)
```

```
model.train(input_fn=input_func,steps=1000)
```

```
pred_fn=tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=len(X_test),shuffle=False)
```

```
pred_gen=model.predict(input_fn=pred_fn)
```

```
predictions=list(pred_gen)
```

```
predictions
```

```
final=[pred['all_class_ids'][0] for pred in predictions]
```
